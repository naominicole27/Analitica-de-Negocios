**CASO DE ESTUDIO**

En este documento se desarrolla y analiza un modelo Naive Bayes para la detección de riesgo SARLAFT, en una entidad financiera utilizando las variables de comportamiento financiero y patrimonial de los clientes. Las variables analizadas son:

edad: Es la edad que posee el cliente al momento de realizar la evaluación de riesgo.

ingresos_usd: Monto total de los ingresos mensuales percibidos por el cliente, expresados en dólares.

gastos_usd: Representa el total de los egresos o gastos mensuales realizados por el cliente en dólares.

num_tarjetas_credito: Indica la cantidad de tarjetas de crédito  que el cliente tiene activos en el sistema financiero.

monto_transado_tarjetas_usd: Es el valor total de las transacciones y consumos realizados por el cliente a través de sus tarjetas de crédito.

porcentaje_crecimiento_patrimonio: Representa la variación porcentual incremental de los bienes y activos del cliente en el último periodo evaluado.

0. Se procede con la carga de la librería de trabajo.

```python
import numpy as np #Librería numérica por excelencia.
import pandas as pd #Librería para la comunicación con archivos excel.

#Librerías específicas.
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix
```

1. Se cargan los datos del trabajo de la base de datos.

```python
nxl="/content/2. LavadoActivos.xlsx"
XDB=pd.read_excel(nxl,sheet_name=0)
XDB.head(100) #Es para mostrar la base de datos cargada.

#Seleccionamos las variables de trabajo.

XD=XDB[['edad', 'ingresos_usd', 'gastos_usd', 'num_tarjetas_credito', 'monto_transado_tarjetas_usd', 'porcentaje_crecimiento_patrimonio']] #Variables de Trabajo
XD.head(10)
yd=XDB['lavado_activos'] #Esta es la variable de decisión.
yd.head(20)
```

2. Implementamos Modelo Naives Bayes

```python
mnb=GaussianNB()
mnb.fit(XD,yd)  #Ajustar Variables Entrada - Salida.

#Mostramos las medias de las variables.

u=mnb.theta_
sigma=mnb.var_; sigma=np.sqrt(sigma)
print("'edad', 'ingresos_usd', 'gastos_usd', 'num_tarjetas_credito', 'monto_transado_tarjetas_usd', 'porcentaje_crecimiento_patrimonio'")
print(u)
print("Las desviaciones son")
print(sigma)
```

**Análisis de Resultados**


De los resultados obtenidos por el modelo podemos observar que la categoría de Riesgo de Lavado (1) posee ingresos superiores a la categoría de Sin Riesgo (0), esto indica claramente la predominancia de la variable ingresos_usd para detectar operaciones sospechosas. Es importante mencionar que el monto transado en tarjetas es significativamente mayor para los perfiles con Riesgo que para los de Sin Riesgo, lo que indica que la entidad debe vigilar los altos volúmenes de transacciones. Podemos decir igualmente que la categoría de Riesgo tiene gastos superiores a los de la categoría Sin Riesgo, guardando relación directa con sus ingresos elevados. Finalmente, variables como el número de tarjetas de crédito, el porcentaje de crecimiento de patrimonio y la edad son muy similares en ambos grupos, lo que sugiere que el modelo no depende de la cantidad de productos o datos demográficos, sino principalmente del flujo de efectivo y la actividad transaccional para la clasificación SARLAFT.
