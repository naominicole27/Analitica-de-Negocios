**CASO DE ESTUDIO**

Se desarrollará un modelo de Árbol de Decisión para predecir el riesgo de enfermedades cardíacas en pacientes, basándose en sus perfiles. El objetivo es clasificar a los individuos en dos estados: Sanos (0) o Enfermos (1), permitiendo así una identificación temprana de riesgos cardiovasculares.
Para que el modelo aprenda a identificar a un paciente en riesgo, utilizaremos las siguientes variables:

Age: Representa la edad cronológica del paciente.

Systolic: Indica la lectura de la presión de la sangre en las arterias cuando el corazón late. Si es muy alta, el corazón se esfuerza de más y puede dañarse.

Diastolic: Indica la lectura de la presión arterial cuando el corazón descansa entre latidos. Es clave para saber si las arterias están siempre bajo mucha tensión.

BMI: Relaciona el peso y la altura del paciente para determinar si existe sobrepeso u obesidad, factores que incrementan el esfuerzo del sistema cardiovascular.

Variable Objetivo:

Disease: Es el resultado final que el modelo busca adivinar.

0 (Sano): El paciente no presenta la enfermedad.

1 (Enfermo): El paciente sí presenta la enfermedad según los datos analizados.

0. Se procede con la carga de las librerías de trabajo.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix
```

1. Se procede con la carga de los datos del trabajo.

```python
nxl='/content/2. HeartDisease.xlsx'
XDB=pd.read_excel(nxl,sheet_name=0)

XD=XDB.iloc[:,[1,2,3,4]]
yd=XDB.iloc[:,0]
display(XDB)
```

2. Se procede con la implementación del módulo árbol.

```python
mar=DecisionTreeClassifier(criterion='gini', max_depth=4)
mar.fit(XD,yd) #Aquí el modelo busca la relación entrada - salida.

#¿Y qué fue lo que hizo el modelo?

ydp=mar.predict(XD)   #Esto es lo que pronosticó el modelo.

#Se construye la matrix de confusion.
cm=confusion_matrix(yd,ydp)

display(cm)
VN=cm[0,0]; FP=cm[0,1]; FN=cm[1,0]; VP=cm[1,1]

#Métricas de desempeño

Ex= (VP+VN)/len(XD)         #1. EXACTITUD: COMPORTAMIENTO GENERAL
print("La exactitud es:", Ex)

Sen=VP/(VP+FN)              #2. SENSIBILIDAD; COMO SE COMPORTA PRONOSTICANDO PREAPROBADOS.
print("La sensibilidad es:", Sen)

Spe=VN/(VN+FP)              #3. ESPECIFICIDAD: COMO SE COMPORTA FRENTE A LOS NEGATIVOS.
print("La especificidad es:", Spe)

Pre=VP/(VP+FP)              #4. PRECISION: COMO SE COMPORTA PRONOSTICANDO PREAPROBADOS
print("La precisión es:",Pre)

PreNeg=VN/(VN+FN)           #5. PRECISION NEGATIVA: COMO SE COMPORTA PRONOSTICANDO NEGATIVOS
print("La precisión negativa es:", PreNeg)
```

**ANÁLISIS DE RESULTADOS**
De la base de datos se puede observar un total de 73 personas que poseen la categoría de Sanos (0), mientras que para la categoría de Enfermos (1) la matriz cuenta con un total de 107 personas. De los 73 datos de personas sanas, el modelo pronosticó correctamente un total de 63 datos (Verdaderos Negativos), mientras que de los 107 enfermos, el modelo pronosticó correctamente un total de 87 datos (Verdaderos Positivos).

El modelo logra identificar en un 86.30% a las personas sanas (63/73), mientras que logró identificar en un 81.31% a las personas enfermas (87/107). Con respecto a las métricas, podemos observar que el modelo logró una Exactitud de 83.33%, lo que indica el buen comportamiento general del modelo frente a la clasificación en las dos categorías de salud. Se destacan la especificidad y la precisión, las cuales lograron valores por encima del 86% y 89% respectivamente, lo que supera ampliamente el límite inferior definido para este tipo de modelos de clasificación, el cual se ubica normalmente en el 75%.

De acuerdo con el árbol de decisión, se destaca la exploración del lado derecho (basado en niveles altos de BMI y Presión Sistólica) donde se encuentran nodos con alta pureza. Se identifica una regla de decisión extrema: SI el paciente posee un BMI > 28.39, una Edad > 45.87 y una Presión Sistólica > 125.58, tendrá una probabilidad de enfermedad del 100% (Nodo puro con Gini = 0). Esto demuestra que el modelo diferencia con gran precisión a los pacientes con perfiles de alto riesgo coronario.

A pesar de la existencia de falsos negativos, se destacan las reglas de negocio para la prevención: el modelo es altamente confiable al emitir diagnósticos positivos, con una Precisión del 89.69%. Esto significa que cuando el sistema clasifica a un sujeto como enfermo, la probabilidad de acierto es casi total, permitiendo priorizar la atención médica en esos casos críticos identificados en el pronóstico final.
